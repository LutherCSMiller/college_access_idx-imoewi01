Second, we tried to give colleges credit only for those lower-income students who graduate rather than all those who enroll.
```python
final = pd.merge(hd, fa, left_on='UNITID',right_on='UNITID')
```

```python
final.fillna(0)
```

```python
final['Net'] = final[['NPT431','NPIS422']].mean(axis=1)
```

```python
final['Net']
```

```python
final = final.sort_values('Net')
```

```python
Chart(final).mark_bar().encode(
                            x=X('Net', axis=Axis(title='Net Cost for Income between 30k and 75k')),
                            y=Y('INSTNM', sort=SortField(field='Net',order='descending')))
```

```python
#individual/average
```

```python
val = final[['Net']].sum()/len(final[['Net']])
final['Ind2'] = final[['Net']]/val
```
